# use enumerate function in loop to build url with monthly dates, and append data to list

```python
# request for 2018 data
precip_2018 = []
end_dates = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31] 

for i, item in enumerate(end_dates):

    response_2018_precip = requests.get(f'{url}key={api_key}&q=94701&format=json&date=2018-{i+1}-01&enddate=2018-{i+1}-{item}&includelocation=yes&tp=24').json()
    print(f"month is {i + 1}")
    
    for i in range(len(response_2018_precip['data']['weather'])):
        print(response_2018_precip['data']['weather'][i]['date'])
        precip_2018.append(float(response_2018_precip['data']['weather'][i]['hourly'][0]['precipMM']))
    print('----------------------------------------')
```
